```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
data = {
    "Temperatura": [20.1, 18.5, 22.3, 19.8, 21.2],
    "Sensores": [101, 110, 95, 98, 105],
    "Refrigeracion": [1, 1, 0, 0, 1]
}
df = pd.DataFrame(data)

# Normalizar los datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Realizar el clustering jerárquico
linked = linkage(scaled_data, method='ward')

# Crear un dendrograma
plt.figure(figsize=(10, 7))
dendrogram(linked,
           orientation='top',
           labels=df.index.tolist(),
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrograma del Clustering Jerárquico')
plt.xlabel('Índice de la muestra')
plt.ylabel('Distancia')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Características
y = np.array([2, 3, 4, 5, 6])  # Variable objetivo

# Inicializar y entrenar el modelo
model = LinearRegression()
model.fit(X, y)

# Predecir valores
X_pred = np.array([6, 7]).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Graficar resultados
plt.scatter(X, y, color='blue')
plt.plot(X_pred, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Datos de ejemplo
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)  # Características
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])  # Variable objetivo

# Inicializar y entrenar el modelo
model = LogisticRegression()
model.fit(X, y)

# Predecir probabilidades
X_pred = np.array([5, 6, 7, 8, 9]).reshape(-1, 1)
y_proba = model.predict_proba(X_pred)[:, 1]

# Graficar resultados
plt.scatter(X, y, color='blue')
plt.plot(X_pred, y_proba, color='red')
plt.xlabel('X')
plt.ylabel('Probabilidad de clase 1')
plt.title('Regresión Logística')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Datos de ejemplo
X = np.array([[0, 0], [1, 1]])
y = np.array([0, 1])

# Inicializar y entrenar el modelo
model = DecisionTreeClassifier()
model.fit(X, y)

# Graficar el árbol de decisión
plt.figure(figsize=(10, 6))
plot_tree(model, filled=True, rounded=True, class_names=['Clase 0', 'Clase 1'], feature_names=['Feature 1', 'Feature 2'])
plt.show()
```
